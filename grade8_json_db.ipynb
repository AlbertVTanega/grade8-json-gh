```python
import firebase_admin
from firebase_admin import credentials, db

# Load the private key
cred = credentials.Certificate("/content/firebase_key.JSON")

# Initialize the app with your database URL
firebase_admin.initialize_app(cred, {
    "databaseURL": "https://grade8-json-db-4079c-default-rtdb.asia-southeast1.firebasedatabase.app/"
})
print("Firebase connected successfully!")
```

```python
import json

with open("/content/students.JSON", "r") as file:
    data = json.load(file)
print("JSON file loaded")
ref = db.reference("students")
for student in data["students"]:
    ref.child(str(student["id"])).set(student)
print("Data uploaded successfully!")
```

```python
students_ref = db.reference("students")
students_data = students_ref.get()
print(students_data)
```

```python
for student_id, info in students_data.items():
    print(info["name"])
```

```python
for student_id, info in students_data.items():
    print(info["name"], info["grades"])
```

```python
# Fetch data from Firebase
students_data = ref.get()
```

```python
for key, student in students_data.items():
    name = student["name"]
    grades = student["grades"].values()
    avg = sum(grades) / len(grades)
    print(f"Student: {name}")
    print(f"Grades: {student['grades']}")
    print(f"Average: {avg:.2f}\n")
```
